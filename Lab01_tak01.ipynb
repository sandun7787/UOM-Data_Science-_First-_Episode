```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have some data to plot, for example, random and standard normal distributions
random_data = np.random.randn(100)
standard_normal_data = np.random.randn(100)

# Plotting the data
plt.plot(random_data, label='Random Distribution')
plt.plot(standard_normal_data, label='Standard Normal Distribution')

# Adding legend, title, and axis labels
plt.legend(loc="upper left")
plt.title("Random and Standard Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

# Display the plot
plt.show()
```

# 3D Plot

```python
import numpy as np
import matplotlib .pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
# Calculate the corresponding Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X**2 + Y**2')

# Add a color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Show the plot
plt.show()

```
